```python
#defining my variables for logistic regression
to_drop=[]
X=ds.drop(['Exited','Surname','CustomerId'],axis=1)
y=ds['Exited']
```

```python
#one hot encoding to accomodate categorical variables
ohe_X=pd.get_dummies(X)
ohe_y=pd.get_dummies(y)
```

```python
#split the dataset into test and train data sets, in a 20-80 ratio respectively.
X_train, X_test, y_train, y_test=train_test_split(ohe_X,ohe_y,test_size=0.2,random_state=42)
```

```python
#buiding the model. 100 decision trees, and 42 as randomness to be able to replicate answers
rf_classifier=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
#training
rf_classifier.fit(X_train,y_train)
```

```python
#test values
# y_pred =rf_classifier.predict(X_test)
#print(type(X_test))
y_pred = rf_classifier.predict(X_test)
```

```python
#evaluate model accuracy
accuracy=accuracy_score(y_test,y_pred)
print("accuracy of model is: " ,accuracy)
```
